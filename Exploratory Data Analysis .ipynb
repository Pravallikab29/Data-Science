```python
#  INTRO TO DATA SCIENCE

# 1 store it into a pandas.DataFramevariable.

import numpy as np
import pandas as pd
titanic_df = pd.read_csv('C:/Users/ABHILASH/Downloads/titanic.csv')   

print(titanic_df.head())             # the given file is stored into titanic_df
print(titanic_df.shape)
print(titanic_df.keys())

print(list(titanic_df.columns))
```

```python
#  to remove all empty cells and count the difference before and after dropping empty cells

# remove empty cells from dataframe and create new dataframe without empty cells
new_df = titanic_df.dropna()
print(new_df.to_string())

# finding no. of rows and columns from new dataframe
print(new_df.shape)
print(titanic_df.shape)

# counting no. of rows/ empty cells removed from original dataframe
print(len(titanic_df) - len(new_df))
```

```python
#  

# removing a label from column and create new dataframe
df =titanic_df.drop(columns=['deck'])
print(df.shape)
new = df.dropna()
print(df.to_string())
#print(new.shape)

# counting no. of empty cells after droping label 'deck' from column
print(len(titanic_df) - len(new))
```

```python
# 

# mean of column 'Fare'
m= new["fare"].mean()
print("Mean",m)

# standard deviation of column 'Fare'
s =new['fare'].std()
print("Standard Deviation", s)

# Maximum of column 'Fare'
a =new['fare'].max()
print("Max of Fare", a)

# Minimum of column 'Fare'
b =new['fare'].min()
print("Minimum of Fare", b)

t1=m+(0.5*s)
t2 = m-(0.5*s)
sata =new[(new['fare']>t1) | (new['fare']<t2)]

print(sata.head())

import matplotlib.pyplot as plt
name =plt.hist(sata['fare'],density=True,bins=30)
plt.ylabel('probability')
plt.xlabel('fare')
plt.show()

```
